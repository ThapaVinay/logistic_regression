## Data Wrangling

```python
titanic_data.isnull()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.drop(labels = "Cabin", axis = 1, inplace = True)
titanic_data.head()
```

```python
titanic_data.dropna(inplace = True)
sns.heatmap(titanic_data.isnull(), cbar = False)
```

```python
Sex = pd.get_dummies(titanic_data["Sex"],drop_first = True)
Embarked = pd.get_dummies(titanic_data["Embarked"],drop_first=True)
Pclass = pd.get_dummies(titanic_data["Pclass"], drop_first = True)
```

```python
titanic_data = pd.concat([titanic_data, Sex, Embarked,Pclass], axis = 1)
titanic_data.head(10)
```

```python
titanic_data.drop(['Pclass', 'Sex', 'Embarked','Name', 'Ticket', 'PassengerId'], axis = 1, inplace = True)
```

```python
titanic_data.head(10)
```

# Train Data

```python
y = titanic_data["Survived"]
X = titanic_data.drop("Survived", axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
# we get converge warning when the dataset is very large and we are not using more number of iterations
# we can scale the data for that
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
X_train
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
